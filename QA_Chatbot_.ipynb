```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
docs = ["""
    Sathyabama Institute of Science and Technology (SIST), formerly known as Sathyabama Engineering College
    and Sathyabama University, is a private,STEM-intensive, multi-disciplinary,
    multi-campus deemed university in Chennai, Tamil Nadu, India. Established in the year 1987 by
    Jeppiaar Jesuadimai, it is a Christian minority educational institution with its patron as Saint Anthony.
    The university main campus is at Sholinganallur, with the dental college nearby at Sithalapakkam and
    a secondary satellite campus at Sriperumbudur.A technical institute that specializes in the engineering fields,
    Sathyabama has been accredited with 'A++' grade by the National Assessment and Accreditation Council (NAAC) and
    Category 1 University by the University Grants Commission (UGC).
    The university is an ISO 9001:2008 certified institution,[6] and in 2016,
    it is built and launched its own space satellite in association with ISRO, India's national space agency.
    Sathyabama has 15 departments that offer 48 accredited undergraduate programs and 23 accredited postgraduate programs
    mostly in the field of engineering, but also in science, technology, law, architecture, medicine, and management.
    The SIST main campus spans across a 140-acres suburban setting located along the IT Corridor.
    The campus buildings include a research hospital, three aeronautical hangars, a science research park, a central library,
    a dental college, and a nanotechnology centre.With 15,600 students, it is one of the largest universities in Tamil Nadu
    """]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
doc_embeddings = model.encode(docs)
```

```python
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(doc_embeddings))
```

```python
synonym_map = {
    "location": "Where is the university located?",
    "place": "Where is Sathyabama University located?",
    "campus": "How many campuses does Sathyabama have?",
    "satellite": "Tell me about the satellite launched by Sathyabama with ISRO.",
    "launch": "Did Sathyabama launch any satellite?",
    "isro": "Tell me about Sathyabama's collaboration with ISRO.",
    "history": "When was Sathyabama Institute established?",
    "founder": "Who founded Sathyabama Institute?",
    "who started": "Who founded Sathyabama Institute?",
    "established": "When was Sathyabama University started?",
    "departments": "How many departments does Sathyabama have?",
    "programs": "What UG and PG programs does Sathyabama offer?",
    "courses": "What courses are available at Sathyabama?",
    "engineering": "Does Sathyabama offer engineering courses?",
    "management": "Does Sathyabama offer management studies?",
    "library": "Tell me about the central library at Sathyabama.",
    "accreditation": "What grade or accreditation has Sathyabama received?",
    "grade": "What NAAC grade does Sathyabama hold?",
    "ugc": "Is Sathyabama recognized by UGC?",
    "naac": "What NAAC grade does Sathyabama have?",
    "hostel": "Does Sathyabama provide hostel facilities?",
    "student count": "How many students study at Sathyabama?",
    "students": "How many students are enrolled?",
    "hospital": "Is there a hospital in the campus?",
    "aeronautics": "Does Sathyabama have aeronautical facilities?",
    "hangar": "Tell me about the aeronautical hangars at Sathyabama.",
    "research": "Tell me about research facilities at Sathyabama.",
    "courses offered": "What programs are offered by Sathyabama?",
    "certification": "Is Sathyabama ISO certified?",
    "facilities": "What are the campus facilities at Sathyabama?",
    "it corridor": "Where is the main campus of Sathyabama located?",
    "private": "Is Sathyabama a private university?",
    "multidisciplinary": "What disciplines does Sathyabama cover?",
    "rank": "What is the university's NAAC or UGC ranking?",
    "education": "What kind of education does Sathyabama provide?",
    "university type": "What type of university is Sathyabama?"
}


query = input("Ask your question: ").strip()
query_lower = query.lower()



if query_lower in synonym_map:
    query = synonym_map[query_lower]


query_vec = model.encode([query])
k = 5
D, I = index.search(query_vec, k)

retrieved_docs = [docs[i] for i in I[0]]
context = "\n".join(retrieved_docs)

print(f"\n Distance to top match: {D[0][0]:.3f}")

if D[0][0] > 1.5:
    print("\n Sorry, I couldn’t find anything clearly related.")
    print(" Try asking about: office locations, services, or customer support.")
else:
        import re
        sentences = re.split(r'(?<=[.!?])\s+', context)

        best_line = max(
          sentences,
          key=lambda s: sum(1 for word in query.lower().split() if word in s.lower()),
          default=retrieved_docs[0]
    )

        print("\n💬 Our best-matching information based on your question:")
        print(f"> {best_line}")


```
